```python
import docx
import pandas as pd
import nltk
from nltk.corpus import twitter_samples
from nltk.tokenize import WhitespaceTokenizer
import random

from nltk import classify
from nltk import NaiveBayesClassifier
```

```python
doc = docx.Document("comments.docx")
all_paras = doc.paragraphs
positive=all_paras[1:50]
negative=all_paras[52:]
```

```python
pos=[]
for i in range(0,49):
    pos.append(positive[i].text)

pos
```

```python
neg=[]
for i in range(0,50):
    neg.append(negative[i].text)

neg
```

```python
nltk.download('punkt')
```

```python
full_dataset=pos+neg
full_dataset
```

```python
tokenizer_input=" ".join(full_dataset)
tokenizer_input
```

```python
tk = WhitespaceTokenizer()

tokens=tk.tokenize(tokenizer_input)

tokens
    

```

```python
def get_tweets_for_model(cleaned_tokens_list):
    for tweet_tokens in cleaned_tokens_list:
        yield dict([token, True] for token in tweet_tokens)
```

```python
positive_tokens_for_model = get_tweets_for_model(pos)
negative_tokens_for_model = get_tweets_for_model(neg)

```

```python
positive_dataset = [(tweet_dict, "Positive")
                     for tweet_dict in positive_tokens_for_model]

negative_dataset = [(tweet_dict, "Negative")
                     for tweet_dict in negative_tokens_for_model]

dataset = positive_dataset + negative_dataset

random.shuffle(dataset)

train_data = dataset[:70]
test_data = dataset[70:]
```

```python
classifier = NaiveBayesClassifier.train(train_data)

print("Accuracy is:", classify.accuracy(classifier, test_data))

print(classifier.show_most_informative_features(10))
```

```python
custom_tweet='Eya niyamai pachayao tikata ridennama denawa.'
custom_tokens=tk.tokenize(custom_tweet)

print(classifier.classify(dict([token, True] for token in custom_tokens)))
```
